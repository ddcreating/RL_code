````python
import numpy as np
import matplotlib.pyplot as plt

# ---------- Random Walk 1000 Environment ----------
class RandomWalk1000:
    """
    States: 0 (terminal, reward=0), 1..1000 (non-terminal), 1001 (terminal, reward=1)
    Start at 500; each step moves left/right with prob 0.5. Gamma=1.0 (episodic).
    """
    def __init__(self, start_state=500, n_states=1000, seed=0):
        assert 1 <= start_state <= n_states
        self.n_states = n_states
        self.start_state = start_state
        self.left_terminal = 0
        self.right_terminal = n_states + 1
        self.rng = np.random.default_rng(seed)
        self.state = start_state

    def reset(self):
        self.state = self.start_state
        return self.state

    def is_terminal(self, s):
        return s == self.left_terminal or s == self.right_terminal

    def step(self, a=None):
        """a is ignored (random walk). Returns (s_next, reward, done)."""
        if self.is_terminal(self.state):
            return self.state, 0.0, True

        # Move left or right with equal probability
        move_right = self.rng.random() < 0.5
        s_next = self.state + (1 if move_right else -1)

        # Rewards only at terminals
        if s_next == self.left_terminal:
            r = 0.0
            done = True
        elif s_next == self.right_terminal:
            r = 1.0
            done = True
        else:
            r = 0.0
            done = False

        self.state = s_next
        return s_next, r, done

# ---------- Feature map & value function ----------
def normalize_state(s, n_states=1000):
    """
    Map state index to [0,1]. Terminals map to 0 and 1 respectively (optional).
    For linear approx we use normalized s for stability.
    """
    if s <= 0:   # left terminal
        return 0.0
    if s >= n_states + 1:  # right terminal
        return 1.0
    return s / (n_states + 1)  # denominator 1001

def phi(s_norm):
    return np.array([s_norm], dtype=float)   # 原来是 [1.0, s_norm]


def v_hat(s, w, n_states=1000):
    return float(w[0] * normalize_state(s, n_states))


def v_star(s, n_states=1000):
    """True value: probability of hitting the right terminal starting from s."""
    if s <= 0: return 0.0
    if s >= n_states + 1: return 1.0
    return s / (n_states + 1)

# ---------- TD(0) update (YOU must fill the update line) ----------
def td0_update(w, s, r, s_next, alpha, n_states=1000):
    gamma = 1.0
    v_s     = v_hat(s, w, n_states)
    v_snext = 0.0 if (s_next == 0 or s_next == n_states + 1) else v_hat(s_next, w, n_states)
    delta = r + gamma * v_snext - v_s
    grad  = phi(normalize_state(s, n_states))   # [1, s_norm]
    w = w + alpha * delta * grad
    return w


# ---------- Training loop ----------
def train_td0(episodes=20, alpha=0.01, seed=0, max_steps_per_episode=50_000):
    env = RandomWalk1000(seed=seed)
    rng = np.random.default_rng(seed)
    w = rng.normal(scale=1e-3, size=2)   # 小随机初始化，而不是全0
    for ep in range(episodes):
        s = env.reset()
        for _ in range(max_steps_per_episode):
            s_next, r, done = env.step()
            w = td0_update(w, s, r, s_next, alpha, n_states=env.n_states)
            s = s_next
            if done:
                break
    return w


# ---------- Evaluate & Plot ----------
def plot_results(w_list, alphas, n_states=1000):
    states = np.arange(1, n_states + 1, 5)
    true_vals = np.array([v_star(s, n_states) for s in states])

    plt.figure(figsize=(8,5))
    plt.plot(states, true_vals, label="True V*(s)=s/1001", linewidth=2)

    for w, a in zip(w_list, alphas):
        approx = np.array([v_hat(s, w, n_states) for s in states])
        plt.plot(states, approx, label=f"Approx (alpha={a})")
    plt.xlabel("state s")
    plt.ylabel("value")
    plt.title("1000-state Random Walk: true vs. learned approximation")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# ---------- Run different alphas ----------
if __name__ == "__main__":
    alphas = [0.001, 0.005, 0.01]  # steps
    w_list = []
    for a in alphas:
        w = train_td0(episodes=50, alpha=a, seed=42, max_steps_per_episode=500000)
        w_list.append(w)
        print(f"alpha={a}: w0={w[0]:.4f}, w1={w[1]:.4f}")

    plot_results(w_list, alphas)
````

## Step 2: SARSA Algorithm with Function Approximation
Implement **SARSA (on-policy TD control)** with the following steps:

For each episode:
1. Initialize state `s`.  
2. Choose action `a` using **ε-greedy** based on Q(s,a).  
3. For each step:  
   - Take action `a`, observe `(s_next, r, done)`.  
   - Choose next action `a_next` using ε-greedy from `s_next`.  
   - Compute TD target:  
     ```
     target = r + gamma * Q(s_next, a_next)
     ```  
     (if `s_next` is terminal, then target = r).  
   - Compute TD error:  
     ```
     delta = target - Q(s,a)
     ```  
   - Update weights:  
     ```
     w <- w + alpha * delta * x(s,a)
     ```  
   - Update `s = s_next`, `a = a_next`.  
4. End episode when the goal is reached or step limit is hit.

````python
# You can use the following code for tiling

import numpy as np


class TileCoderXY:
    """
    A TileCoder for function approximation that applies tile coding on the x and y coordinates
    of a 3D state. Instead of providing tile widths, the user provides the number of tiles per
    dimension. The tile widths are computed based on the state bounds and the number of tiles.
    The z coordinate is not used.
    """

    def __init__(self, num_tilings, tiles_per_dim, state_low, state_high):
        """
        Initialize the TileCoderXY.

        Parameters:
            num_tilings (int): Number of overlapping tilings.
            tiles_per_dim (array-like of 2 ints): Number of tiles along the x and y dimensions.
            state_low (array-like of 2 floats): Lower bounds for the x and y dimensions.
            state_high (array-like of 2 floats): Upper bounds for the x and y dimensions.
        """
        self.num_tilings = num_tilings
        self.tiles_per_dim = np.array(tiles_per_dim, dtype=int)
        self.state_low = np.array(state_low, dtype=float)
        self.state_high = np.array(state_high, dtype=float)

        # Compute the tile width for each dimension.
        # We assume that the grid spans exactly from state_low to state_high.
        # When there are N tiles, there are N-1 intervals between the boundaries.
        self.tile_width = (self.state_high - self.state_low) / (self.tiles_per_dim - 1)

        # Precompute an offset for each tiling to create overlapping grids.
        # self.offsets = [(i / self.num_tilings) * self.tile_width for i in range(self.num_tilings)]
        # self.offsets = self.compute_8_offsets()
        # self.offsets = np.stack(self._compute_offsets(), axis=0)  # shape: (num_tilings, dims)
        # Precompute offsets for each tiling.
        # For tiling i:
        #   offset_x = (((i + 0) % num_tilings) / num_tilings) * tile_width[0]
        #   offset_y = (((i + 1) % num_tilings) / num_tilings) * tile_width[1]
        offsets = np.empty((self.num_tilings, 2))
        for i in range(self.num_tilings):
            offsets[i, 0] = (((i + 0) % self.num_tilings) / self.num_tilings) * self.tile_width[0]
            offsets[i, 1] = (((i + 1) % self.num_tilings) / self.num_tilings) * self.tile_width[1]
        self.offsets = offsets


        # Precompute multiplier for flattening a 2D index.
        # For grid shape (N, M), flat index = x_index * M + y_index.
        self.multiplier = self.tiles_per_dim[1]

        # Initialize a weight vector for each tiling.
        num_tiles = np.prod(self.tiles_per_dim)
        self.weights = [np.zeros(num_tiles) for _ in range(self.num_tilings)]

    def save(self, file_name):
        np.savez(file_name + ".npz", weights=self.weights)

    def load(self, file_name):
        self.weights = np.load(file_name+".npz")["weights"]


    def compute_8_offsets(self):
        """
        Compute a list of offsets using a combination of cardinal and diagonal directions.
        The offsets include:
          - Center: [0, 0]
          - Cardinal: right, left, up, down (half-tile shifts)
          - Diagonal: up-right, up-left, down-right, down-left (half-tile shifts)

        If the number of tilings exceeds the number of unique offsets (9), the list is repeated.

        Returns:
            List of 2-element numpy arrays representing the offset for each tiling.
        """
        half_tile = self.tile_width / 8.0
        base_offsets = [
            np.array([0.0, 0.0]),  # Center (no shift)
            np.array([half_tile[0], 0.0]),  # Right
            np.array([-half_tile[0], 0.0]),  # Left
            np.array([0.0, half_tile[1]]),  # Up
            np.array([0.0, -half_tile[1]]),  # Down
            np.array([half_tile[0], half_tile[1]]),  # Up-right
            np.array([-half_tile[0], half_tile[1]]),  # Up-left
            np.array([half_tile[0], -half_tile[1]]),  # Down-right
            np.array([-half_tile[0], -half_tile[1]])  # Down-left
        ]
        offsets = []
        for i in range(self.num_tilings):
            offsets.append(base_offsets[i % len(base_offsets)])
        return offsets

    def get_tile_indices(self, state):
        """
        Compute the active tile indices for all tilings given a 2D state.

        Parameters:
            state (array-like of length 2): The input state [x, y].

        Returns:
            List of tuples (tiling_index, flat_tile_index) for each tiling.
        """
        state = np.array(state, dtype=float)  # shape: (2,)
        # Compute shifted states for all tilings in one vectorized operation.
        # Shape of shifted: (num_tilings, 2)
        shifted = (state - self.state_low) + self.offsets

        # Compute tile coordinates (integer indices) for each tiling.
        # Division is broadcasted over the offsets.
        tile_coords = (shifted / self.tile_width).astype(int)  # shape: (num_tilings, 2)

        # Clip to ensure indices are within bounds.
        tile_coords[:, 0] = np.clip(tile_coords[:, 0], 0, self.tiles_per_dim[0] - 1)
        tile_coords[:, 1] = np.clip(tile_coords[:, 1], 0, self.tiles_per_dim[1] - 1)

        # Compute flat indices for each tiling.
        # flat_index = x_index * (tiles_per_dim[1]) + y_index
        flat_indices = tile_coords[:, 0] * self.tiles_per_dim[1] + tile_coords[:, 1]

        # Return a list of (tiling_index, flat_index) tuples.
        return list(zip(range(self.num_tilings), flat_indices))


    def predict(self, state):
        """
        Compute the approximated function value for a given 3D state using tile coding on x and y.

        Parameters:
            state (array-like): The input state [x, y, z].

        Returns:
            float: The function approximation (sum of weights for the active tiles).
        """
        active_tiles = self.get_tile_indices(state)
        return sum(self.weights[tiling][idx] for tiling, idx in active_tiles)

    def update(self, state, target, alpha):
        """
        Update the weights given a state and target value.

        Parameters:
            state (array-like): The input state [x, y, z].
            target (float): The target function value.
            alpha (float): The overall learning rate.
        """
        prediction = self.predict(state)
        error = target - prediction
        # Distribute the learning rate equally among all tilings.
        alpha_per_tiling = alpha / self.num_tilings

        active_tiles = self.get_tile_indices(state)
        for tiling, idx in active_tiles:
            self.weights[tiling][idx] += alpha_per_tiling * error



````

````python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
````

````python
class SARSAAgent:
    def __init__(self, num_actions, num_tilings=8, tiles_per_dim=[8, 8],
                 state_low=[-1.2, -0.07], state_high=[0.6, 0.07],
                 alpha=0.3, gamma=0.99, epsilon=0.1):
        """
        Initialize SARSA agent with tile coding function approximation.

        Args:
            num_actions: Number of discrete actions
            num_tilings: Number of overlapping tilings
            tiles_per_dim: Tiles along each dimension [x, y]
            state_low: Lower bounds for state space
            state_high: Upper bounds for state space
            alpha: Learning rate
            gamma: Discount factor
            epsilon: Exploration rate for epsilon-greedy
        """
        self.num_actions = num_actions
        self.alpha = alpha
        self.gamma = gamma
        self.epsilon = epsilon

        # Create separate tile coder for each action
        self.tile_coders = [
            TileCoderXY(num_tilings, tiles_per_dim, state_low, state_high)
            for _ in range(num_actions)
        ]

    def get_q_values(self, state):
        """Get Q-values for all actions in given state."""
        return np.array([coder.predict(state) for coder in self.tile_coders])

    def epsilon_greedy(self, state):
        """Select action using epsilon-greedy policy."""
        if np.random.random() < self.epsilon:
            return np.random.randint(self.num_actions)
        else:
            return np.argmax(self.get_q_values(state))

    def update(self, state, action, target):
        """Update Q-function for given state-action pair."""
        self.tile_coders[action].update(state, target, self.alpha)

    def train_episode(self, env):
        """Run one episode of SARSA training."""
        state, _ = env.reset()
        action = self.epsilon_greedy(state)
        total_reward = 0

        while True:
            # Take action and observe next state
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            total_reward += reward

            # Choose next action using epsilon-greedy
            next_action = self.epsilon_greedy(next_state)

            # Compute SARSA target
            if done:
                target = reward
            else:
                q_next = self.tile_coders[next_action].predict(next_state)
                target = reward + self.gamma * q_next

            # Update Q-function
            self.update(state, action, target)

            if done:
                break

            # Move to next state-action pair
            state = next_state
            action = next_action

        return total_reward
````

````python
def run_random_policy(env, num_episodes=10):
    """Run episodes with random policy as baseline."""
    returns = []
    for _ in range(num_episodes):
        state, _ = env.reset()
        total_reward = 0
        done = False

        while not done:
            action = env.action_space.sample()
            state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            total_reward += reward

        returns.append(total_reward)

    return returns



def train_sarsa(num_episodes=1000):
    """Train SARSA agent on MountainCar environment."""
    env = gym.make('MountainCar-v0')
    agent = SARSAAgent(
        num_actions=env.action_space.n,
        num_tilings=8,
        tiles_per_dim=[8, 8],
        alpha=0.3,
        gamma=0.99,
        epsilon=0.1
    )

    returns = []
    for episode in range(num_episodes):
        episode_return = agent.train_episode(env)
        returns.append(episode_return)

        if (episode + 1) % 100 == 0:
            avg_return = np.mean(returns[-100:])
            print(f"Episode {episode + 1}, Avg Return (last 100): {avg_return:.2f}")

    env.close()
    return agent, returns
````

````python
def plot_results(sarsa_returns, random_returns):
    """Plot training results and compare with random policy."""
    plt.figure(figsize=(12, 5))

    # Plot episode returns
    plt.subplot(1, 2, 1)
    plt.plot(sarsa_returns, alpha=0.3, label='SARSA')
    # Smooth curve with moving average
    window = 50
    smoothed = np.convolve(sarsa_returns, np.ones(window)/window, mode='valid')
    plt.plot(range(window-1, len(sarsa_returns)), smoothed,
             linewidth=2, label='SARSA (smoothed)')
    plt.axhline(np.mean(random_returns), color='r', linestyle='--',
                label=f'Random Policy (avg={np.mean(random_returns):.1f})')
    plt.xlabel('Episode')
    plt.ylabel('Episode Return')
    plt.title('Learning Curve: SARSA vs Random Policy')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Plot performance comparison
    plt.subplot(1, 2, 2)
    final_performance = sarsa_returns[-100:]
    plt.boxplot([random_returns, final_performance],
                labels=['Random', 'SARSA (final 100)'])
    plt.ylabel('Episode Return')
    plt.title('Performance Comparison')
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()


def plot_value_function(agent, resolution=50):
    """Visualize the learned value function V(s) = max_a Q(s,a)."""
    # Create grid over state space
    positions = np.linspace(-1.2, 0.6, resolution)
    velocities = np.linspace(-0.07, 0.07, resolution)

    # Compute value function
    V = np.zeros((resolution, resolution))
    for i, pos in enumerate(positions):
        for j, vel in enumerate(velocities):
            state = [pos, vel]
            q_values = agent.get_q_values(state)
            V[j, i] = np.max(q_values)

    # Plot value function
    plt.figure(figsize=(10, 8))

    # 2D heatmap
    plt.subplot(2, 1, 1)
    plt.imshow(V, extent=[-1.2, 0.6, -0.07, 0.07],
               aspect='auto', origin='lower', cmap='viridis')
    plt.colorbar(label='Value V(s)')
    plt.xlabel('Position')
    plt.ylabel('Velocity')
    plt.title('Value Function Heatmap')
    plt.axvline(0.5, color='r', linestyle='--', label='Goal')
    plt.legend()

    # 3D surface plot
    ax = plt.subplot(2, 1, 2, projection='3d')
    X, Y = np.meshgrid(positions, velocities)
    surf = ax.plot_surface(X, Y, V, cmap='viridis', alpha=0.8)
    ax.set_xlabel('Position')
    ax.set_ylabel('Velocity')
    ax.set_zlabel('Value')
    ax.set_title('Value Function (3D)')
    plt.colorbar(surf, ax=ax, shrink=0.5)

    plt.tight_layout()
    plt.show()
````

````python
# Step 1: Run random policy baseline
print("=" * 60)
print("MountainCar-v0 with SARSA and Tile Coding")
print("=" * 60)

print("\n[Step 1] Running random policy baseline...")
env = gym.make('MountainCar-v0')
random_returns = run_random_policy(env, num_episodes=50)
env.close()
print(f"Random policy average return: {np.mean(random_returns):.2f}")

# Step 2: Train SARSA agent
print("\n[Step 2] Training SARSA agent...")
agent, sarsa_returns = train_sarsa(num_episodes=1000)
print(f"SARSA final 100 episodes average: {np.mean(sarsa_returns[-100:]):.2f}")
````

````python
print("\n[Step 3] Generating visualizations...")
plot_results(sarsa_returns, random_returns)
````

## Discussion

The SARSA agent successfully learned to solve the MountainCar problem, improving from -184 to approximately -146 average return by episode 300. This represents a 27% improvement over the random policy baseline (-200), demonstrating that the agent learned the momentum-building strategy required to reach the goal. The learning curve shows rapid initial improvement followed by stable performance, indicating effective convergence.

The tile coding parameters (8 tilings, 8×8 resolution) provided a good balance for this task. Increasing the number of tilings would improve generalization and produce smoother value functions but slow convergence, while fewer tilings would learn faster but with coarser approximation. Higher resolution (e.g., 16×16) would enable finer state discrimination but require more training episodes, whereas lower resolution (e.g., 4×4) would converge faster but miss important state distinctions. The current configuration achieved stable performance within 300 episodes, suggesting it is well-suited for the MountainCar environment.
